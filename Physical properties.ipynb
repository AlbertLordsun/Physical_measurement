```python
# locomotion trajectory
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

data = pd.read_csv(r'')
dis = data['v']; t = np.arange(0.01,3.01,0.01)

plt.figure(0, figsize=(8,10))
for i in range(12):
    plt.plot(t, 0.4*(dis[:300,2*i]+11-i), color = cm.jet((11-i)/12), linewidth=4)

plt.xlabel("Time (sec)", fontsize=30)
plt.ylabel("Displacement (mm)", fontsize=30)
plt.xlim([0,3.1])
plt.ylim([-0.5,5])
plt.xticks([0,1,2,3])
plt.yticks([0,1,2,3,4,5])
plt.tick_params(labelsize=24)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.legend()
plt.savefig("Kymograph-no friction.png", facecolor="white", bbox_inches = "tight")
    
plt.show()
```

```python
# Plot for the tension trajectory

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r'D:\Research files\Nose kohsaka lab\Soft robotics\Design\Matlab\Neuromechanical model\MyModel\NM\Data\Larvae measurement\20200310\ATR+Gal4-1.csv')
x = np.array(data['sec']); y = np.array(data['N']);

plt.figure(figsize = (10.5,6))
plt.plot(x[1000:2000]-50, y[1000:2000]+0.001, color = 'y', linewidth = 3)
plt.xlabel('Time (sec)'); plt.ylabel('Force (N)');
plt.rcParams.update({'font.size': 21})
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300
plt.savefig('Tension trajectory.png',dpi = 600)
plt.show()
```
